## **EXAMPLE 4.2: Grouping and Aggregate functions**

Sometimes feeling cold is more than about low temperatures; a windy day can also make you cold
A factor called the chill factor can be used to quantify the cold based on the wind speed and the temperature. The formula for chill factor is given by

WCI = (10 * sqroot(v) -v + 10.5).(33-Tm)

v is the speed of the wind and Tm is the minimum temperature

Add a column for WCI and find the month with the lowest WCI

```python
from math import sqrt
def wci(x):
  velocity = x["WindGustSpeed"]
  minTemp = x["MinTemp"]
  return (10*sqrt(velocity)-velocity+10.5)*(33-minTemp)
```

```python
data['WCI'] = data.apply(wci, axis = 1)
data.head()
```

```python
data_byMonth = data.groupby(by=["Month"]).mean()
data_byMonth.head()
```

```python
data_byMonth.sort_values("WCI", ascending=False).head()
```
